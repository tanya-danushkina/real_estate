# Исследование объявлений о продаже квартир


В нашем распоряжении данные сервиса Н. — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. По каждой квартире на продажу доступны два вида данных: 
- вписанные пользователем
- полученные автоматически на основе картографических данных (расстояние до центра, аэропорта, ближайшего парка и водоёма) 

**Задача**: выполнить предобработку данных и изучить их, для нахождения зависимостей на рынке недвижимости.

**План работы**: 
1. Обзор данных
2. Предобработка данных
3. Добавление новых параметров (цена одного квадратного метра, день недели публикации объявления, месяц публикации объявления, год публикации объявления, тип этажа квартиры, расстояние до центра города в километрах)
4. Исследовательский анализ
5. Общий вывод

**Описание данных**:<br>
`airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)<br>
`balcony` — число балконов<br>
`ceiling_height` — высота потолков (м)<br>
`cityCenters_nearest` — расстояние до центра города (м)<br>
`days_exposition` — сколько дней было размещено объявление (от публикации до снятия)<br>
`first_day_exposition` — дата публикации<br>
`floor` — этаж<br>
`floors_total` — всего этажей в доме<br>
`is_apartment` — апартаменты (булев тип)<br>
`kitchen_area` — площадь кухни в квадратных метрах (м²)<br>
`last_price` — цена на момент снятия с публикации<br>
`living_area` — жилая площадь в квадратных метрах (м²)<br>
`locality_name` — название населённого пункта<br>
`open_plan` — свободная планировка (булев тип)<br>
`parks_around3000` — число парков в радиусе 3 км<br>
`parks_nearest` — расстояние до ближайшего парка (м)<br>
`ponds_around3000` — число водоёмов в радиусе 3 км<br>
`ponds_nearest` — расстояние до ближайшего водоёма (м)<br>
`rooms` — число комнат<br>
`studio` — квартира-студия (булев тип)<br>
`total_area` — общая площадь квартиры в квадратных метрах (м²)<br>
`total_images` — число фотографий квартиры в объявлении<br>

## Слобец `rooms`

Рассмотрим столбец `rooms`, проанализируем уникальные значения. 

```python
print('Количество квартир по числу комнат')
df['rooms'].value_counts()
```

В датасете есть 10-19-комнатные квартиры, скорее всего это ошибка в данных. Кваритр не очень много, можно их удалить. Проверим, отражается ли такое число комнат на площади:

```python
df.query('rooms >=10')['total_area'].sort_values()
```

Все эти квартиры с большой площадью, пока оставим их в датасете. 

Кроме того, остаются квартиры, где 0 комнат, такого быть не может, рассмотрим площади этих квартир: 

```python
df.query('rooms == 0')['total_area'].describe()
```

Средняя площать таких квартир – 29м^2, это похоже на однокомнатную квартиру (в среднем площать однокомнатной квартиры до 39м^2) или студию. Проверим сколько квартир, где площадь больше 39м^2:

```python
df.query('rooms == 0 and total_area > 39')['total_area'].value_counts()
```

Можно предположить, что квартиры до 39м^2 однокомнатные, от 40 до 60 – двухкомнатные и от 61 до 85 – трёхкомнатные. Все остальные можно удалить, так как сложно сказать, сколько там комнат. 

```python
df.loc[(df['rooms']==0) & (df['total_area']<40), 'rooms'] = 1
df.loc[(df['rooms']==0) & (df['total_area']>39) & (df['total_area']<61), 'rooms'] = 2
df.loc[(df['rooms']==0) & (df['total_area']>59) & (df['total_area']<86), 'rooms'] = 3

df.drop(df.loc[(df['rooms']==0) & (df['total_area']>86)].index,inplace=True, axis=0)

df['rooms'].value_counts()
```

**Вывод:** <br>
Столбец `rooms` – корректный.

## Столбец `ceiling height`

В столбце присутствует 39% пропусков. Можно предположить, что высота потолков в таких квартирах стандартная: можно заполнить медианным значением (медиана более устойчива к выбросам, чем средннее) по количеству этажей в доме.

```python
for ceiling in df['floors_total'].unique():
    median = df.loc[df['floors_total'] == ceiling, 'ceiling_height'].median()
    df.loc[(df['ceiling_height'].isna()) & (df['floors_total'] == ceiling), 'ceiling_height'] = median
    
df.loc[df['ceiling_height'].isna(), 'ceiling_height'] = df['ceiling_height'].median()
print('Кол-во пропусков в столбце `is_apartment`', df['ceiling_height'].isna().sum())
```

Проверим данные на выбросы: 

```python
df.boxplot(column=['ceiling_height'], color='pink')

plt.title('Выбросы в данных (высота потолков)')
plt.xlabel('Количество')
plt.ylabel('Высота потолка')

plt.show()
```

```python
index_list = df.query('ceiling_height > 19 and ceiling_height < 40').index 
df.loc[index_list, 'ceiling_height'] = df.loc[index_list, 'ceiling_height'] / 10
print(df.loc[index_list]['ceiling_height'])
```

```python
df.boxplot(column=['ceiling_height'], color='pink')

plt.title('Выбросы в данных (высота потолков)')
plt.xlabel('Количество')
plt.ylabel('Высота потолка')

plt.show()
```

```python
df.query('ceiling_height < 2 or ceiling_height > 4.5')
```

Выбросов осталось не так много, поэтому от них можно избавиться (высота потолка не может быть ниже 2м и выше 4.5м). 

```python
df = df[(df['ceiling_height'] > 2) & (df['ceiling_height'] < 4.5)]
df.info()
```

```python
df.boxplot(column=['ceiling_height'], color='pink')

plt.title('Выбросы в данных (высота потолков)')
plt.xlabel('Количество')
plt.ylabel('Высота потолка')

plt.show()
```

**Вывод:** <br>
Столбец `ceiling_height` – корректный.

## Столбец `living_area`

Зависимость площади от количества комнат высокая, поэтому логично заполнить пропуски в столце средней по кол-ву комнат в квартире: 

```python
mean_living_area_by_rooms = df.groupby('rooms')['living_area'].transform('mean')
df['living_area'] = df['living_area'].fillna(mean_living_area_by_rooms)
print('Кол-во пропусков в столбце `living_area`', df['living_area'].isna().sum())
```

**Вывод:** <br>
Столбец `living_area` – корректный.

## Столбец `is_apartment`

## Распределение квартир по признакам

Построим гистограммы по значениям за вычетом выбросов:
